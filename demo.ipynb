# MNIST from Scratch - Interactive Demo

This notebook demonstrates:
- Loading the MNIST dataset
- Visualizing handwritten digit samples
- Analyzing the data distribution

Run the cells below to explore the data!

```python
# Load MNIST data
print("Loading MNIST dataset...")
(train_images, train_labels), (test_images, test_labels) = load_mnist()

print(f"Training data shape: {train_images.shape}")
print(f"Training labels shape: {train_labels.shape}")
print(f"Test data shape: {test_images.shape}")
print(f"Test labels shape: {test_labels.shape}")

# Show some statistics
print(f"\nUnique labels in training set: {np.unique(train_labels)}")
print(f"Sample counts per digit:")
for i in range(10):
    count = np.sum(train_labels == i)
    print(f"  Digit {i}: {count} samples")
```

```python
# Visualize some training samples
print("Visualizing training samples...")

# Set up matplotlib for inline display
%matplotlib inline
plt.style.use('default')

# Create a figure with subplots
fig, axes = plt.subplots(2, 5, figsize=(12, 6))
axes = axes.ravel()

# Show first 10 training samples
for i in range(10):
    axes[i].imshow(train_images[i], cmap='gray')
    axes[i].set_title(f'Label: {train_labels[i]}', fontsize=12)
    axes[i].axis('off')

plt.suptitle('MNIST Training Samples', fontsize=16)
plt.tight_layout()
plt.show()

print("These are handwritten digits from the MNIST dataset!")
print("Each image is 28x28 pixels, normalized to [0,1]")
```

```python
# Show distribution of digits
print("Analyzing digit distribution...")

# Count samples per digit
digit_counts = [np.sum(train_labels == i) for i in range(10)]

# Create bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(range(10), digit_counts, color='skyblue', edgecolor='black')
plt.xlabel('Digit', fontsize=12)
plt.ylabel('Number of Samples', fontsize=12)
plt.title('Distribution of Digits in MNIST Training Set', fontsize=16)
plt.xticks(range(10))

# Add value labels on bars
for bar, count in zip(bars, digit_counts):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 100,
             f'{count}', ha='center', va='bottom', fontsize=10)

plt.grid(axis='y', alpha=0.3)
plt.show()

print("The dataset is fairly balanced across all digits!")
print(f"Total training samples: {len(train_labels)}")
print(f"Average samples per digit: {np.mean(digit_counts):.0f}")
```
